```python
# Use this cell to begin your analyses, and add as many cells as you would like!
import pandas as pd

lego_sets = pd.read_csv('datasets/lego_sets.csv',header=0,parse_dates=['year'])
parent_themes = pd.read_csv('datasets/parent_themes.csv')
```

```python
print(lego_sets.head())
print(lego_sets.dtypes)

print(parent_themes.head())
print(parent_themes.dtypes)
```

```python
lego_sets= lego_sets.dropna(subset=['set_num'])
```

```python
# Confirm if any other column contains missing value
# In this case it is ok for num_parts.
lego_sets.isnull().any()
```

```python
parent_themes.isnull().any()
```

```python
lego = lego_sets.merge(parent_themes,left_on='parent_theme', right_on ='name', how='left')
```

```python
lego.head()
```

```python
lego.columns = ['set_num', 'name', 'year','num_parts','theme_name','parent_theme','parent_theme_id', 'parent_theme_','is_licensed']
```

```python
# test if there is one set_num belonging to multiple parent_theme
test_set_parent =lego[['set_num','parent_theme_id']].drop_duplicates().groupby('set_num').count() 
assert (test_set_parent>1).any()[0] == False
```
